# Part 2: Plotting

```octave
fprintf('Plotting Data ... \n')
data = load('assignment/mlclass-ex1/ex1data1.txt');
X = data(:, 1); y = data(:, 2);
m = length(y);
plotData(X, y)

hold on;
ylabel('Profit in $10,000s')
xlabel('Population of City in 10,000s')
hold off;
```

# Part 3: Gradient descent

```octave
X = [ones(m, 1), data(:, 1)]; % add a column of ones to x
theta = zeros(2, 1);

% some GD settings
iterations = 1500;
alpha = 0.01;

computeCost(X, y, theta)
theta = gradientDescent(X, y, theta, alpha, iterations)
fprintf('Theta found by GD: \n');
fprintf('%f %f \n', theta(1), theta(2));
```

## Plot the linear fit

```octave
plotData(data(:, 1), y)
hold on;
plot(X(:, 2), X*theta, '-')
ylabel('Profit in $10,000s')
xlabel('Population of City in 10,000s')
legend('Training data', 'Linear regression')
hold off;
```

```octave
predict1 = [1, 3.5] * theta;
predict2 = [1, 7] * theta;
fprintf('For population = 35,000, predict profit %f\n', predict1 * 10000);
fprintf('For population = 70,000, predict profit %f\n', predict2 * 10000);
```

# Part 4: Visualizing $J(\theta_0, \theta_1)$

```octave
theta0_vals = linspace(-10, 10, 100);
theta1_vals = linspace(-1, 4, 100);

J_vals = zeros(length(theta0_vals), length(theta1_vals));
for i = 1:length(theta0_vals)
    for j=1:length(theta1_vals);
        t = [theta0_vals(i); theta1_vals(j)];
        J_vals(i, j) = computeCost(X, y, t);
    end
end
```

according to the document how meshgrids work
`J_vals` is the Z axis matrix and the mapping are
- X maps to columns of Z
- Y maps to rows of Z

therefore we have to transpose our `J_vals`

```octave
J_vals = J_vals';
```

## 3D surface plot

```octave
figure;
surf(theta0_vals, theta1_vals, J_vals);
xlabel('\theta_0'); ylabel('\theta_1');
```

## contour plot

```octave
figure;
contour(theta0_vals, theta1_vals, J_vals, logspace(-2, 3, 25))
hold on;
xlabel('\theta_0'); ylabel('\theta_1');
plot(theta(1), theta(2), 'rx', 'MarkerSize', 10, 'LineWidth', 2);
hold off
```
